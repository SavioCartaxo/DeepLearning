```python
ds = ds.drop(columns=["PoolQC", "MiscFeature", "Fence", "Alley"])

ds = pd.get_dummies(ds, columns=["MSZoning"])
ds = pd.get_dummies(ds, columns=["Street"])
ds = pd.get_dummies(ds, columns=["LotShape"])

LandContour_map = {
    "Lvl" : 0,
    "HLS" : 1,
    "Bnk" : 2,
    "Low" : 3,
}

ds["LandContour"] = ds["LandContour"].map(LandContour_map)

Utilities_map = {
    "AllPub" : 0,
    "NoSewr" : 1,
    "NoSeWa" : 2,
    "ELO" : 3
}

ds["Utilities"] = ds["Utilities"].map(Utilities_map)

ds = pd.get_dummies(ds, columns=["LotConfig"])
ds = pd.get_dummies(ds, columns=["LandSlope"])
ds = pd.get_dummies(ds, columns=["Neighborhood", "Condition1", "Condition2", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl", "Exterior1st", "Exterior2nd", "MasVnrType", ])

ExterQual_map = {
    "Ex" : 5,
    "Gd" : 4,
    "TA" : 3,
    "Fa" : 2,
    "PO" : 1,
}

ds["ExterQual"] = ds["ExterQual"].map(ExterQual_map)
ds["ExterCond"] = ds["ExterCond"].map(ExterQual_map)

ds = pd.get_dummies(ds, colums=["Foundation", ])

BsmtQual_map = {
    "Ex" : 5,
    "Gd" : 4,
    "TA" : 3,
    "Fa" : 2,
    "PO" : 1,
    "NA" : 0
}

ds["BsmtQual"] = ds["BsmtQual"].map(BsmtQual_map)
ds["BsmtCond"] = ds["BsmtCond"].map(BsmtQual_map)

BsmtExposure_map = {
    "Gd" : 4,
    "Av" : 3,
    "Mn" : 2,
    "No" : 1,
    "NA" : 0,
}

ds["BsmtExposure"] = ds["BsmtExposure"].map(BsmtExposure_map)

BsmtFinType1_map = {
    "GLQ": 6,
    "ALQ": 5,
    "BLQ": 4,
    "Rec": 3,
    "LwQ": 2,
    "Unf": 1,
    np.nan: 0
}

ds["BsmtFinType1"] = ds["BsmtFinType1"].map(BsmtFinType1_map)

BsmtFinType2_map = {
    "GLQ": 6,
    "ALQ": 5,
    "BLQ": 4,
    "Rec": 3,
    "LwQ": 2,
    "Unf": 1,
    "NA": 0
}

ds["BsmtFinType2"] = ds["BsmtFinType2"].map(BsmtFinType2_map)

HeatingQC_map = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1,
    "NA": 0
}

ds["HeatingQC"] = ds["HeatingQC"].map(HeatingQC_map)

KitchenQual_map = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1
}

ds["KitchenQual"] = ds["KitchenQual"].map(KitchenQual_map)

Functional_map = {
    "Typ": 7,
    "Min1": 6,
    "Min2": 5,
    "Mod": 4,
    "Maj1": 3,
    "Maj2": 2,
    "Sev": 1,
    "Sal": 0
}

ds["Functional"] = ds["Functional"].map(Functional_map)

FireplaceQu_map = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1,
    "NA": 0
}

ds["FireplaceQu"] = ds["FireplaceQu"].map(FireplaceQu_map)

GarageFinish_map = {
    "Fin": 3,
    "RFn": 2,
    "Unf": 1,
    "NA": 0
}

ds["GarageFinish"] = ds["GarageFinish"].map(GarageFinish_map)

GarageQual_map = {
    "Ex": 5,
    "Gd": 4,
    "TA": 3,
    "Fa": 2,
    "Po": 1,
    "NA": 0
}

ds["GarageQual"] = ds["GarageQual"].map(GarageQual_map)
ds["GarageCond"] = ds["GarageCond"].map(GarageQual_map)

PoolQC_map = {
    "Ex": 4,
    "Gd": 3,
    "TA": 2,
    "Fa": 1,
    "NA": 0
}

ds["PoolQC"] = ds["PoolQC"].map(PoolQC_map)

Fence_map = {
    "GdPrv": 4,
    "MnPrv": 3,
    "GdWo": 2,
    "MnWw": 1,
    "NA": 0
}

ds["Fence"] = ds["Fence"].map(Fence_map)

ds["CentralAir"] = ds["CentralAir"].map({"Y": 1, "N": 0})

ds["PavedDrive"] = ds["PavedDrive"].map({
    "Y": 2,
    "P": 1,
    "N": 0
})

ds = pd.get_dummies(ds, columns=["Heating", "Electrical", "GarageType", "MiscFeature", "SaleType", "SaleCondition"])

```
